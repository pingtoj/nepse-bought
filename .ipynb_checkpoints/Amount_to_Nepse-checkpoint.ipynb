```python
def SubIndex(debenture_,mutual_,ordinary_,shl_):
    i,j,k,l = 0,0,0,0
    for index, row in debenture_.iterrows():
        i += row['Open Price'] * 25
    for index, row in mutual_.iterrows():
        j += row['Open Price'] * 100
    for index, row in ordinary_.iterrows():
        k += row['Open Price'] * 10 
    for index, row in shl_.iterrows():
        l += row['Open Price'] * 100 
    
    return i,j,k,l
```
