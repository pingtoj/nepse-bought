```python
def total(mutual_,ordinary_,shl_):
    i,j,k = 0,0,0
    for index, row in mutual_.iterrows():
        i += row['Close Price'] * 100
    for index, row in ordinary_.iterrows():
        j += row['Close Price'] * 10 
    for index, row in shl_.iterrows():
        k += row['Close Price'] * 100 
    
    return i,j,k
```

```python
total(mutual_,ordinary_,shl_)
```

```python
net_total=sum(total(mutual_,ordinary_,shl_))
```
